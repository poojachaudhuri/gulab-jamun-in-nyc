```python
# get the id, name, rating, coordinates, price of all the businesses

for business in business_data['businesses']:
    if 'price' not in business:
        print(business['id'], business['name'], business['rating'], business['location']['zip_code'], business['coordinates']['latitude'], business['coordinates']['longitude'], 'N/A')
    else:
        print(business['id'], business['name'], business['rating'], business['location']['zip_code'], business['coordinates']['latitude'], business['coordinates']['longitude'], business['price'])
```

```python
businesses = []

for business in business_data['businesses']:
    if 'price' not in business:
        businesses.append({'id':business['id'], 'name':business['name'], 'rating':business['rating'], 'zip':business['location']['zip_code'], 'lat':business['coordinates']['latitude'], 'long':business['coordinates']['longitude'], 'price':'N/A'})
    else:
        businesses.append({'id':business['id'], 'name':business['name'], 'rating':business['rating'], 'zip':business['location']['zip_code'], 'lat':business['coordinates']['latitude'], 'long':business['coordinates']['longitude'], 'price':business['price']})
```
