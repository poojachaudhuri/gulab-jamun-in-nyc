### Matcha cheesecake 

There are 169 results for matcha cheesecake. Yelp API has an offset parameter that can be used to get more results. 

```python


matcha_params = {"term":["matcha cheesecake", "green tea cheesecake"], "location":"new york city", "limit": 50}
results = []
offset = 0

while len(results) < 760:
    matcha_params["offset"] = offset
    response = requests.get(url=endpoint, headers=headers, params=matcha_params)
    matcha_data = response.json()
    matcha_businesses = matcha_data["businesses"]
    results += matcha_businesses
    offset += 50

results = results[:760]
results
```

I chose 760 because it threw up KeyError: Businesses for more so I assumed that there is no 'businesses' key beyond that i.e., there are no results beyond 760. I saved the output into a CSV file. Now when I run the code it throws up KeyError for any number of results. I read online that Yelp API sometimes misbehaves. I assume this is what's happening. I am going ahead with the data I was able to get.

```python
# make a new column 'category' that categorizes the restaurants as cheap, medium, or expensive

price_df['category'] = price_df['price'].apply(lambda x: 'Cheap' if x < 5 else ('Medium' if x < 8 else 'Expensive'))
price_df.head()

```

```python
# make a bar plot of the number of restaurants in each category


price_df['category'].value_counts().plot(kind='bar')
plt.xticks(rotation=0)
plt.yticks(np.arange(0, 25, 5))
plt.title('Restaurants categorized by price', fontdict={'fontsize': 12, 'fontweight': 5, 'color': 'brown'})
plt.ylabel('Number of restaurants', fontdict={'fontsize': 10, 'fontweight': 5, 'color': 'brown'})
for index, value in enumerate(price_df['category'].value_counts()):
    plt.text(index, value, str(value), ha='center', va='bottom')
# annotate the graph
plt.annotate("Cheap < 5 dollars\nMedium = 6 to 8 dollars \nExpensive > 8 dollars", xy=(0, 16), xytext=(1.4, 16))
```
